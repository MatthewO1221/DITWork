```python
#import libraries

import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
#load csv file

#use load text to get the data and labels separately

iris_data = np.loadtxt(open('iris.csv', 'rb'), delimiter=',', skiprows=1, usecols=(0,1,2,3), dtype=None, encoding='utf-8')

iris_labels = np.loadtxt(open('iris.csv', 'rb'), delimiter=',', skiprows=1, usecols=4, dtype=str, encoding='utf-8')


print(iris_data.shape)
print(iris_labels.shape)

#combine into one array
iris = np.column_stack((iris_data, iris_labels))

print(iris.shape)

print(iris)
```

```python
#Put data pieces in their own arrays to make life easier


sepal_length = iris[:, 0].astype(float)
sepal_width = iris[:, 1].astype(float)
petal_length = iris[:, 2].astype(float)
petal_width = iris[:, 3].astype(float)
labels = iris[:, 4]

# Plot histograms for each data piece

#Sepal length plot
plt.subplot(2, 2, 1)
plt.hist(sepal_length, bins=10, color='blue')
plt.title('Sepal Length')

#Sepal width plot
plt.subplot(2, 2, 2)
plt.hist(sepal_width, bins=10, color='green')
plt.title('Sepal Width')

#Petal length plot
plt.subplot(2, 2, 3)
plt.hist(petal_length, bins=10, color='red')
plt.title('Petal Length')

#Petal width plot
plt.subplot(2, 2, 4)
plt.hist(petal_width, bins=10, color='purple')
plt.title('Petal Width')

#Had to use chatgpt for this, makes it so text doesn't overlap
plt.tight_layout()
plt.show()

#Print minimum, maximum, and average values
print("Sepal Length - Min:", np.min(sepal_length), "Max:", np.max(sepal_length), "Average:", np.mean(sepal_length))
print("Sepal Width - Min:", np.min(sepal_width), "Max:", np.max(sepal_width), "Average:", np.mean(sepal_width))
print("Petal Length - Min:", np.min(petal_length), "Max:", np.max(petal_length), "Average:", np.mean(petal_length))
print("Petal Width - Min:", np.min(petal_width), "Max:", np.max(petal_width), "Average:", np.mean(petal_width))
```
