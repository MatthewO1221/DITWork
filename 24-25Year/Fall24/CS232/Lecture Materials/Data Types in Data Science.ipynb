## Data Types

Using Pandas the possibility is opened to use different types, not only numerals.
<br>
A column can be considered a feature or variable. It has values (observations) in each row.

The major two different types are:
**Numerical** and **Categorical**<br>
These are also called: **Quantitave** and **Qualitative**

Numerical can be: **Discrete** or **Continuous**<br>
Categorical can be: **Nominal** or **Ordinal** or **Binary**

**Numerical**: measurable information, represented by numbers (not strings)
- Continuous: the values have no logical sense (but measure a feature with precision), usually floats
    - Examples:
        - weight
        - height
        - temperature
- Discrete: the values have logical sense (related to each other and/or representative), usually integers
    - Examples:
        - 1, 2, 3, 4, 5, 6, 7, where the numbers stand for days of the week
        - number of courses taken in a semester
        - number of days until St. Patrick Day
        - number of bugs in your game project

**Categorical**: categories described using strings
- Ordinal: the values have a set/established order between them
    - Examples:
        - {"Sunday", "Monday", "Tuesday", "Wednesday", ...}
        - {"January", "February", "March", ...}
        - {"A+", "A", "B+", "B", "B-", ...}
        - {"Critical", "Normal", "Low"}

- Nominal: the values **do not** have a set/established order between them
    - Examples:
        - {"Seattle", "Boston", "Miami", "Chicago"}
        - {"Al", "Ben", "Cindy"}
        - {"Single", "Married"}
- Binary: two distinct values
    - Examples:
        - {"this", "that"}
        - {Yes, No}
        - {True, False}
        - {1, 0} - note that these numbers describe categories
